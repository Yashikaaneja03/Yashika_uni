```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/agriculture.csv')
df['state'] = df['state'].str.replace(',','').str.replace('yr','').str.split('-').str[0]

df.plot(x='state',kind='bar')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.hist(df['min_price'], bins=10)
plt.xlabel('max_price')
plt.ylabel('Frequency')
plt.title('agriculture')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.hist(df['min_price'], bins=10)
plt.xlabel('district')
plt.ylabel('state')
plt.title('agriculture')
plt.show()
```
